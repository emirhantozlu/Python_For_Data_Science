```python
def cat_summary(dataframe, col_name):
    summary_df = pd.DataFrame({
        col_name: dataframe[col_name].value_counts(),
        "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)
    })
    print(summary_df)
    print("_______________________________________________________________")

cat_summary(df, "sex")
```

## Kategorik Değişken Analizi 2

```python
def cat_summary(dataframe, col_name,plot=False):
    summary_df = pd.DataFrame({
        col_name: dataframe[col_name].value_counts(),
        "Ratio": 100 * dataframe[col_name].value_counts() / len(dataframe)
    })
    print(summary_df)
    print("_______________________________________________________________")

    if plot:
        sns.countplot(x=dataframe[col_name],data=dataframe)
        plt.show(block=True)

```

```python
cat_summary(df,"sex",True)
```

```python
for col in cat_cols:
    plt.title(col)
    if df[col].dtypes=="bool":
        print("görselleştirme yapılamadı")
    else:
        cat_summary(df,col,True)
```

```python
for col in cat_cols:
    plt.title(col)  
    if df[col].dtypes == "bool":  
        df[col] = df[col].astype(int)
    else:
        cat_summary(df, col, True)  
    plt.show()  
```

## Değişkenlerin Yakalanması

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
# docstring senin yazdığın fonksiyonu başkasının anlaması için açılama yazmaktır.
def grab_col_names(dataframe,cat_th=10,car_th=20):
    """
        veri setindeki kategorik, numerik ve kategorik fakat kardinal değişkenlerin isimlerini verir.

        Parametre
        -----------------
        dataframe: dataframe
            değişken isimleri alınmak istenen dataframe'dir.
        cat_th: int,float
            numerik fakat kategorik olan değişkenler için sınıf eşik değeridir.
        car_th: int,float
            kategorik fakat kardinal değişkenler için sınıf eşik değeridir.

        returns
        -----------------
        cat_cols: list
            kategorik değişken listesi
        num_cols: list
            numerik değişken listesi
        cat_but_car: list
            kategorik görünümlü kardinal değişken listesi

        notes
        -----------------
        cat_cols + num_cols + cat_but_car = toplam değişken sayısı
        num_but_cat cat_cols'un içerisinde.
    """
    cat_cols=[col for col in df.columns if str(df[col].dtypes) in ["category","object","bool"]]
    
    cat_but_cat=[col for col in df.columns if df[col].nunique() > 10 and str(df[col].dtypes) in ["int64","float64"]]
    
    num_but_car=[col for col in df.columns if 
                df[col].nunique() > 20 and str(df[col].dtypes) in ["category","object"]]
    
    cat_cols=cat_cols+num_but_cat
    
    cat_cols=[col for col in cat_cols if col not in cat_but_car]


    num_cols=[col for col in df.columns if df[col].dtypes in ["int64","float64"]]
    num_cols=[col for col in num_cols if col not in cat_cols]

    print(f"Observations: {dataframe.shape[0]}")
    print(f"Variables : {dataframe.shape[1]}")
    print(f"cat_cols: {len(cat_cols)}")
    print(f"num_cols: {len(num_cols)}")
    print(f"cat_but_car: {len(cat_but_car)}")
    print(f"num_but_cat: {len(num_but_cat)}")

    return cat_cols,num_cols,cat_but_car
    

cat_cols,num_cols,cat_but_car=grab_col_names(df)

    
```

```python
for col in cat_cols:
    cat_summary(df,col)
    
```

```python
for col in num_cols:
    num_summary(df,col,plot=True)
```

```python
# bonus (bool türündeki değişkenleri int'e çevirdik)
df=sns.load_dataset("titanic")
for col in df.columns:
    if df[col].dtypes=="bool":
        df[col]=df[col].astype(int)

cat_cols,num_cols,cat_but_car=grab_col_names(df)
```

```python
for col in num_cols:
    num_summary(df,col,plot=True)
```

```python
def target_summary_width_num(dataframe, target, numerical_col):
    # Hedef değişkeni gruplayarak numerik sütunun ortalamasını alıyoruz
    print(pd.DataFrame({col: dataframe.groupby(target)[numerical_col].mean()}))
    print("-------------------------------------------------------------------")

target_summary_width_num(df,"survived","age")
```

```python
def high_correlated_cols(dataframe, plot=False, corr_th=0.90):
    # Sadece sayısal sütunları seç
    numeric_df = dataframe.select_dtypes(include=[np.number])
    
    # Korelasyon matrisi oluşturuluyor
    corr = numeric_df.corr()
    cor_matrix = corr.abs()
    
    # Üst üçgen matris elde ediliyor
    upper_triangle_matrix = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))
    
    # Belirlenen threshold'a göre yüksek korelasyonlu sütunlar belirleniyor
    drop_list = [col for col in upper_triangle_matrix.columns if any(upper_triangle_matrix[col] > corr_th)]
    
    # Eğer plot=True ise heatmap çiziliyor
    if plot:
        import seaborn as sns
        import matplotlib.pyplot as plt
        sns.set(rc={"figure.figsize": (15, 15)})
        sns.heatmap(corr, cmap="RdBu", annot=True)
        plt.show()
    
    # drop_list döndürülüyor
    return drop_list

# Örnek çağırma
high_correlated_cols(df)
```

```python
drop_list=high_correlated_cols(df, plot=True)
```

```python
df.drop(drop_list,axis=1)
```

```python
high_correlated_cols(df.drop(drop_list,axis=1),plot=True)
```
