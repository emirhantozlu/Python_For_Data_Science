# Python İle Veri Analizi: Veri Görselleştirme 

## Kategorik Değişken Görselleştirme

```python
# kategorik bir değişken varsa bunu sütun grafiği ile görselleştirilmelidir.
# countplot bar
```

```python
# sayısal bir değişken varsa hist veya boxplot kullanılmalıdır.
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns",None) # üç çizgi ile göstermesin istedik
pd.set_option("display.width",500)
df=sns.load_dataset("titanic")
df.head()
```

```python
df["sex"].value_counts()
```

```python
df["sex"].value_counts().plot(kind="bar")
plt.show()
```

## Sayısal Değişken Görselleştirme

```python
plt.hist(df["age"])
plt.show()
```

```python
plt.boxplot(df["fare"])
plt.show()
```

## Matplotlip Özellikleri

```python
# plot

x = np.array([1, 8])
y = np.array([0, 150])

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, "o")
plt.show()
```

```python
a = np.array([2, 4, 6, 8, 10])
b = np.array([1, 3, 5, 7, 9])

plt.plot(a, b)
plt.show()
```

```python
a = np.array([2, 4, 6, 8, 10])
b = np.array([1, 3, 5, 7, 9])

plt.plot(a, b, "o")
plt.show()
```

```python
# marker

c = np.array([13, 28, 11, 100])

plt.plot(c, marker="o")
plt.show()
```

```python
c = np.array([13, 28, 11, 100])

plt.plot(c, marker="*")
plt.show()
```

```python
# line
c = np.array([13, 28, 11, 100])

plt.plot(c, linestyle="dashed")
plt.show()
```

```python
plt.plot(c, linestyle="dotted")
plt.show()
```

```python
plt.plot(c, linestyle="dashdot")
plt.show()
```

```python
plt.plot(c, linestyle="dashdot",color="r")
plt.show()
```

```python
# multple lines
```

```python
x=np.array([23,18,31,10])
y=np.array([13,28,11,100])

plt.plot(x)
plt.plot(y)
plt.show()
```

```python
## labels
```

```python
x=np.array([23,18,31,10])
y=np.array([13,28,11,100])
plt.plot(x,y)
plt.title("ana başlık")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.grid() # ızgara ekledik
plt.show()
```

```python

x = np.array([23, 18, 31, 10, 23, 65, 89, 54, 62, 35, 98])
y = np.array([13, 28, 11, 100, 154, 268, 236, 147, 23, 65, 35])

# Subplot 1: 1. grafik
plt.subplot(1, 2, 1)  # 1 satır, 2 sütun, 1. grafik
plt.title("Grafik 1")
plt.plot(x, y)

# Subplot 2: 2. grafik
plt.subplot(1, 2, 2)  # 1 satır, 2 sütun, 2. grafik
plt.title("Grafik 2")
plt.plot(y, x)
plt.show()
```

## Seaborn İle Veri Görselleştirme

```python
df=sns.load_dataset("tips")
df.head()
```

```python
df["sex"].value_counts()
```

```python
sns.countplot(x=df["sex"],data=df)
plt.show()
```
