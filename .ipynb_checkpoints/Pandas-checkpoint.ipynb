## Toplulaştırma & Gruplama

```python
df=sns.load_dataset("titanic")

df["age"].mean()
```

```python
df.groupby("sex")["age"].mean()
```

```python
df.groupby("sex").agg({"age":"mean"})
```

```python
df.groupby("sex").agg({"age":["mean","sum"]})
```

```python
df.groupby("sex").agg({"age":["mean","sum"],
                       "survived":"mean"})
```

```python
df.groupby(["sex","embark_town"]).agg({"age":["mean"]
                ,"survived" : "mean"})
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age":["mean"]
                ,"survived" : "mean"})
```

```python
df.groupby(["sex","embark_town","class"]).agg({"age":["mean"]
                ,"survived" : "mean", "sex":"count"  })
```

## Apply & Lambda

```python
df=sns.load_dataset("titanic")
# apply satır ya da sütünlarda otomatik olarak fonksiyon çalıştırmamızı sağlar.
# lambda kullan at fonksiyon tanımlamamızı sağlar.
```

```python
df["age2"]=df["age"]*2
df["age3"]=df["age"]*5
```

```python
for col in df.columns:
    if "age" in col:
        df[col]=df[col]/10

df.head()
```

```python
df[["age","age2","age3"]].apply(lambda x:x/10).head()
```

```python
df.loc[:,df.columns.str.contains("age")].apply(lambda x:x/10).head()
# öncelikle tüm satırları gezdik ve içinde age olanları seçtik daha sonra fonksiyonumuzu uyguladık.
```

```python
df.loc[:,df.columns.str.contains("age")].apply(lambda x: (x-x.mean()) / (x.std())).head()
```

```python
def standart_scaler(col_name):
    return (col_name-col_name.mean()) / col_name.std()

df.loc[:,df.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
df.loc[: , df.columns.str.contains("age")]=df.loc[: , df.columns.str.contains("age")].apply(standart_scaler)
df.head()
```

## Birleştirme İşlemleri

```python
import numpy as np
m=np.random.randint(1,30,size=(5,3))
df1=pd.DataFrame(m,columns=["var1","var2","var3"])
df2=df1+99
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
# merge ile birleştirme işlemleri
```

```python
df1=pd.DataFrame({"employees":["jhon","denis","mark","maria"],
                 "group":["accounting","engineering","engineering","hr"]})

df2=pd.DataFrame({"employees":["jhon","denis","mark","maria"],
                 "start_date":[2010,2009,2014,2019]})
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on="employees") 
# hangi değişkene göre birleştirmesi gerektiğini biz seçtik.
```

```python
df3=pd.merge(df1,df2)
df4=pd.DataFrame({"group":["accounting","engineering","hr"],
                 "manager" : ["caner","mustafa","berkcan"]})
```

```python
pd.merge(df3,df4)
```
